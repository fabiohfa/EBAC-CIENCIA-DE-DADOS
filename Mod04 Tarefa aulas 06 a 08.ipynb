####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
def consulta_url():
    url_uf,simb,url_em,texto_uf, interesses = '','','',"de todo o Brasil", "Estados e Municípios"
    print("Por favor, insira a sigla da Unidade da Federação que deseja consultar (Caso queira a consulta todos os valores apenas teclar 'enter')")
    uf = str(input())
    print("Insira o tipo de ente da federação que deseja consultar Estado ou Município (Caso queira a consulta todos os valores apenas teclar 'enter')")
    estado_municipio = str(input())
    estado_municipio = estado_municipio.capitalize()
    if uf != "":
        url_uf, texto_uf = f'uf={uf}', f'{uf}'.upper()
    if estado_municipio !="":
        url_em,interesses = f'tipo_interessado={estado_municipio}', f'{estado_municipio}'
    if (estado_municipio !="") & (uf != ""):
        simb = "&"
    r = requests.get(f"https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?{url_uf}{simb}{url_em}")
    dataframe = pd.DataFrame(r.json()['items'])
    print(f'Você selecionou os dados {texto_uf} e o tipo de dado engloba {interesses}.')
    return dataframe
````

````python
# 1) Versão sem input
def consulta_url2(uf="",estado_municipio=""):
    url_uf,simb,url_em,texto_uf, interesses = '','','',"de todo o Brasil", "Estados e Municípios"
    estado_municipio = estado_municipio.capitalize()
    if uf != "":
        url_uf, texto_uf = f'uf={uf}', f'{uf}'.upper()
    if estado_municipio !="":
        url_em,interesses = f'tipo_interessado={estado_municipio}', f'{estado_municipio}'
    if (estado_municipio !="") & (uf != ""):
        simb = "&"
    r = requests.get(f"https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?{url_uf}{simb}{url_em}")
    dataframe = pd.DataFrame(r.json()['items'])
    print(f'Você selecionou os dados {texto_uf} e o tipo de dado engloba {interesses}.')
    return dataframe
````

````python
consulta_url2(uf="RS")
````
